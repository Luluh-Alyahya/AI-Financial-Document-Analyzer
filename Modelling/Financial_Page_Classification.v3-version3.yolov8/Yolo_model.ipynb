```python
from ultralytics import YOLO
import pandas as pd
import os, glob

model = YOLO(r"best.pt")

source_dir = r"C:\Users\Lolo-\OneDrive\سطح المكتب\Sukuk Assighnment\Modelling\Financial_Page_Classification.v3-version3.yolov8\test\images"
images = glob.glob(os.path.join(source_dir, "*.jpg")) + glob.glob(os.path.join(source_dir, "*.png"))

print(f"Found {len(images)} images to process...")

batch_size = 8
predictions = []

for i in range(0, len(images), batch_size):
    batch = images[i:i+batch_size]
    print(f"Processing batch {i//batch_size + 1} of {len(images)//batch_size + 1} ...")

    results = model.predict(
        source=batch,
        imgsz=640,      
        conf=0.2,
        save=False,
        verbose=False
    )

    for r in results:
        image_name = os.path.basename(r.path)
        labels = [r.names[int(b.cls)] for b in r.boxes]
        unique_labels = sorted(set(labels))
        predictions.append({
            "image": image_name,
            "labels_found": ", ".join(unique_labels) if unique_labels else "None"
        })

df = pd.DataFrame(predictions)
df.to_csv("yolo_predictions_fast.csv", index=False)
print(f" Done! Saved results for {len(df)} images to yolo_predictions_fast.csv")
```

##### For pages classification

```python
import pandas as pd

df = pd.read_csv("yolo_predictions_grouped.csv")

def classify(labels):
    labels = {lbl.strip().lower() for lbl in str(labels).split(",")}
    if "auditor_report" in labels:
        return "auditor report page"
    if "financial_sheets" in labels or ("financial_sheets" in labels and "table" in labels):
        return "financial sheets page"
    if "notes" in labels and "text" in labels:
        return "text notes page"
    if "notes" in labels and "table" in labels:
        return "tabular notes page"
    if "table"  in labels:
        return "financial sheets page"
    return "other page"

df["classified"] = df["labels_found"].apply(classify)
df.to_csv("classified_pages.csv", index=False)
print("Classification done and saved to classified_pages.csv")
```
